#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
ds_precios_positivos = ds[ds['price'] > 0]

# Encontrar la propiedad más barata y seleccionar columnas
casa_mas_barata = ds_precios_positivos.loc[ds_precios_positivos['price'].idxmax(), ['address', 'country', 'price']]

# Mostrar resultados
print("Información de la propiedad más cara:")
print("Dirección:", casa_mas_barata['address'])
print("País:", casa_mas_barata['country'])
print("Precio:", casa_mas_barata['price'], "€")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# Filtrar propiedades con precio mayor que 0
ds_precios_positivos = ds[ds['price'] > 0]

# Encontrar la propiedad más barata y seleccionar columnas
casa_mas_barata = ds_precios_positivos.loc[ds_precios_positivos['price'].idxmin(), ['address', 'country', 'price']]

# Mostrar resultados
print("Información de la propiedad más barata:")
print("Dirección:", casa_mas_barata['address'])
print("País:", casa_mas_barata['country'])
print("Precio:", casa_mas_barata['price'], "€")
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

```python
# Filtrar propiedades con superficie válida (mayor que 0 y no nula)
ds_superficie_valida = ds[(ds['surface'] > 0) & (ds['surface'].notna())]

# Casa más grande
casa_mas_grande = ds_superficie_valida.loc[ds_superficie_valida['surface'].idxmax()]
# Casa más pequeña
casa_mas_pequena = ds_superficie_valida.loc[ds_superficie_valida['surface'].idxmin()]

# Mostrar resultados
print("🏠 CASA MÁS GRANDE 🏠")
print("Superficie:", casa_mas_grande['surface'], "m²")
print("Dirección:", casa_mas_grande['address'])
print("País:", casa_mas_grande['country'])

print("\n🏠 CASA MÁS PEQUEÑA 🏠")
print("Superficie:", casa_mas_pequena['surface'], "m²")
print("Dirección:", casa_mas_pequena['address'])
print("País:", casa_mas_pequena['country'])
```

#### Ejercicio 06. Elimina los NAs del dataset, si aplica (★★☆)

Imprima una comparación entre las dimensiones del DataFrame original versus el DataFrame después de las eliminaciones.


```python
# Guardar dimensiones originales
dim_original = ds.shape

# Eliminar filas con NA (todos los valores faltantes)
ds_sin_na = ds.dropna()

# Guardar nuevas dimensiones
dim_sin_na = ds_sin_na.shape

# Crear DataFrame comparativo
comparacion = pd.DataFrame({
    'Dataset': ['Original', 'Sin NAs'],
    'Filas': [dim_original[0], dim_sin_na[0]],
    'Columnas': [dim_original[1], dim_sin_na[1]],
    'Filas eliminadas': [0, dim_original[0] - dim_sin_na[0]]
})

# Mostrar resultados
print("Comparación de dimensiones:")
print(comparacion.to_string(index=False))

# Mostrar porcentaje de datos perdidos
porcentaje_perdido = ((dim_original[0] - dim_sin_na[0]) / dim_original[0]) * 100
print(f"\nSe eliminó el {porcentaje_perdido:.2f}% de las filas originales")
```

#### Ejercicio 07. ¿Cuál la media de precios en la población (columna level5) de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido.

```python
## Filtrar propiedades en Arroyomolinos (sin NA y precio > 0)
filtro = (
    ds['level4'].str.contains('Arroyomolinos', case=False, na=False) & 
    (ds['price'] > 0)
)
arroyomolinos_data = ds[filtro]

# Calcular estadísticas
if not arroyomolinos_data.empty:
    media_precio = arroyomolinos_data['price'].mean()
    cantidad_propiedades = len(arroyomolinos_data)
    
    print(f"🔍 Análisis para: Arroyomolinos (Madrid) (excluyendo precio 0)")
    print(f"• Propiedades válidas encontradas: {cantidad_propiedades}")
    print(f"• Precio promedio: {media_precio:,.2f} €")
    print(f"• Rango de precios válidos: {arroyomolinos_data['price'].min():,.2f} € - {arroyomolinos_data['price'].max():,.2f} €")
    
    # Propiedades más económicas (excluyendo 0)
    print("\n💸 3 propiedades más económicas (con precio > 0):")
    print(arroyomolinos_data.nsmallest(3, 'price')[['address', 'price', 'surface', 'rooms']].to_string(index=False))
    
else:
    print("No se encontraron propiedades válidas (precio > 0) en Arroyomolinos (Madrid)")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población (level5 column) de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
# Filtrar propiedades en Arroyomolinos con precio > 0
filtro = (
    ds['level4'].str.contains('Arroyomolinos', case=False, na=False) & 
    (ds['price'] > 0)
)
arroyomolinos_data = ds[filtro]

# Crear histograma
plt.figure(figsize=(12, 6))
plt.hist(arroyomolinos_data['price'], 
         bins=20, 
         color='#4CAF50', 
         edgecolor='darkgreen',
         alpha=0.7)

# Personalización
plt.title('Distribución de Precios en Arroyomolinos (Madrid)', fontsize=15, pad=20)
plt.xlabel('Precio (€)', fontsize=12)
plt.ylabel('Número de Propiedades', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir línea de media
media = arroyomolinos_data['price'].mean()
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, 
            label=f'Media: {media:,.2f} €')

# Formatear ejes
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f} €'))
plt.xticks(rotation=45)

plt.legend()
plt.tight_layout()
plt.show()

# Estadísticas complementarias
print(f"\n📊 Estadísticas clave de precios en Arroyomolinos (n={len(arroyomolinos_data)} propiedades):")
print(f"• Media: {media:,.2f} €")
print(f"• Mediana: {arroyomolinos_data['price'].median():,.2f} €")
print(f"• Rango: {arroyomolinos_data['price'].min():,.2f} € - {arroyomolinos_data['price'].max():,.2f} €")
```
