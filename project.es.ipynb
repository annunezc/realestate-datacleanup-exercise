#### Ejercicio 08. Trazar el histograma de los precios para la poblaci√≥n (level5 column) de "Arroyomolinos (Madrid)" y explica qu√© observas (‚òÖ‚òÖ‚òÜ)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve an√°lisis del trazado.


```python
import matplotlib.pyplot as plt
# Filtrar propiedades en Arroyomolinos con precio > 0
filtro = (
    ds['level4'].str.contains('Arroyomolinos', case=False, na=False) & 
    (ds['price'] > 0)
)
arroyomolinos_data = ds[filtro]

# Crear histograma
plt.figure(figsize=(12, 6))
plt.hist(arroyomolinos_data['price'], 
         bins=20, 
         color='#4CAF50', 
         edgecolor='darkgreen',
         alpha=0.7)

# Personalizaci√≥n
plt.title('Distribuci√≥n de Precios en Arroyomolinos (Madrid)', fontsize=15, pad=20)
plt.xlabel('Precio (‚Ç¨)', fontsize=12)
plt.ylabel('N√∫mero de Propiedades', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# A√±adir l√≠nea de media
media = arroyomolinos_data['price'].mean()
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, 
            label=f'Media: {media:,.2f} ‚Ç¨')

# Formatear ejes
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f} ‚Ç¨'))
plt.xticks(rotation=45)

plt.legend()
plt.tight_layout()
plt.show()

# Estad√≠sticas complementarias
print(f"\nüìä Estad√≠sticas clave de precios en Arroyomolinos (n={len(arroyomolinos_data)} propiedades):")
print(f"‚Ä¢ Media: {media:,.2f} ‚Ç¨")
print(f"‚Ä¢ Mediana: {arroyomolinos_data['price'].median():,.2f} ‚Ç¨")
print(f"‚Ä¢ Rango: {arroyomolinos_data['price'].min():,.2f} ‚Ç¨ - {arroyomolinos_data['price'].max():,.2f} ‚Ç¨")
```

podemos ver en el grafico que la media de los precios de las casas en Arroyomolinos es de 299.000 euros aproximadamente. Ademas podemos observar que hay muy pocas casas con valores mayores a 400.000 euros

#### Ejercicio 10. ¬øSon los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (‚òÖ‚òÖ‚òÜ)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusi√≥n sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
ds['pps'] = ds['price'] / ds['surface']  # Crear columna precio/m¬≤

# Filtrar datos v√°lidos para ambas localidades
filtro = (
    (ds['pps'].notna()) & 
    (ds['pps'] != float('inf')) & 
    (ds['price'] > 0) & 
    (ds['surface'] > 0)
)

valdemorillo = ds[filtro & ds['level4'].str.contains('Valdemorillo', case=False, na=False)]
galapagar = ds[filtro & ds['level4'].str.contains('Galapagar', case=False, na=False)]

# Calcular promedios
media_valdemorillo = valdemorillo['pps'].mean()
media_galapagar = galapagar['pps'].mean()

# Resultados
print("üí∂ PROMEDIO PRECIO POR M¬≤:")
print(f"‚Ä¢ Valdemorillo: {media_valdemorillo:,.2f} ‚Ç¨/m¬≤ (n={len(valdemorillo)} propiedades)")
print(f"‚Ä¢ Galapagar:    {media_galapagar:,.2f} ‚Ç¨/m¬≤ (n={len(galapagar)} propiedades)")
print(f"‚Ä¢ Diferencia:   {abs(media_valdemorillo - media_galapagar):,.2f} ‚Ç¨/m¬≤")
```
