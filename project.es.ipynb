#### Ejercicio 07. ¿Cuál la media de precios en la población (columna level5) de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido.

```python
## Filtrar propiedades en Arroyomolinos (sin NA y precio > 0)
filtro = (
    ds['level4'].str.contains('Arroyomolinos', case=False, na=False) & 
    (ds['price'] > 0)
)
arroyomolinos_data = ds[filtro]

# Calcular estadísticas
if not arroyomolinos_data.empty:
    media_precio = arroyomolinos_data['price'].mean()
    cantidad_propiedades = len(arroyomolinos_data)
    
    print(f"🔍 Análisis para: Arroyomolinos (Madrid) (excluyendo precio 0)")
    print(f"• Propiedades válidas encontradas: {cantidad_propiedades}")
    print(f"• Precio promedio: {media_precio:,.2f} €")
    print(f"• Rango de precios válidos: {arroyomolinos_data['price'].min():,.2f} € - {arroyomolinos_data['price'].max():,.2f} €")
    
    # Propiedades más económicas (excluyendo 0)
    print("\n💸 3 propiedades más económicas (con precio > 0):")
    print(arroyomolinos_data.nsmallest(3, 'price')[['address', 'price', 'surface', 'rooms']].to_string(index=False))
    
else:
    print("No se encontraron propiedades válidas (precio > 0) en Arroyomolinos (Madrid)")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población (level5 column) de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
# Filtrar propiedades en Arroyomolinos con precio > 0
filtro = (
    ds['level4'].str.contains('Arroyomolinos', case=False, na=False) & 
    (ds['price'] > 0)
)
arroyomolinos_data = ds[filtro]

# Crear histograma
plt.figure(figsize=(12, 6))
plt.hist(arroyomolinos_data['price'], 
         bins=20, 
         color='#4CAF50', 
         edgecolor='darkgreen',
         alpha=0.7)

# Personalización
plt.title('Distribución de Precios en Arroyomolinos (Madrid)', fontsize=15, pad=20)
plt.xlabel('Precio (€)', fontsize=12)
plt.ylabel('Número de Propiedades', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir línea de media
media = arroyomolinos_data['price'].mean()
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, 
            label=f'Media: {media:,.2f} €')

# Formatear ejes
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f} €'))
plt.xticks(rotation=45)

plt.legend()
plt.tight_layout()
plt.show()

# Estadísticas complementarias
print(f"\n📊 Estadísticas clave de precios en Arroyomolinos (n={len(arroyomolinos_data)} propiedades):")
print(f"• Media: {media:,.2f} €")
print(f"• Mediana: {arroyomolinos_data['price'].median():,.2f} €")
print(f"• Rango: {arroyomolinos_data['price'].min():,.2f} € - {arroyomolinos_data['price'].max():,.2f} €")
```

#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
# Filtrar propiedades válidas (precio > 0) en ambas localidades
filtro_valdemorillo = (ds['level4'].str.contains('Valdemorillo', case=False, na=False) & (ds['price'] > 0))
filtro_galapagar = (ds['level4'].str.contains('Galapagar', case=False, na=False) & (ds['price'] > 0))

valdemorillo_data = ds[filtro_valdemorillo]
galapagar_data = ds[filtro_galapagar]

# Calcular estadísticas
media_valdemorillo = valdemorillo_data['price'].mean()
media_galapagar = galapagar_data['price'].mean()

# Crear tabla comparativa
comparacion = pd.DataFrame({
    'Localidad': ['Valdemorillo', 'Galapagar'],
    'Precio Promedio (€)': [media_valdemorillo, media_galapagar],
    'Propiedades Analizadas': [len(valdemorillo_data), len(galapagar_data)],
    'Diferencia con Valdemorillo': [0, media_galapagar - media_valdemorillo]
})

print("📊 Comparación de Precios Promedio:")
print(comparacion.to_string(index=False))



```

#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
ds['pps'] = ds['price'] / ds['surface']  # Crear columna precio/m²

# Filtrar datos válidos para ambas localidades
filtro = (
    (ds['pps'].notna()) & 
    (ds['pps'] != float('inf')) & 
    (ds['price'] > 0) & 
    (ds['surface'] > 0)
)

valdemorillo = ds[filtro & ds['level4'].str.contains('Valdemorillo', case=False, na=False)]
galapagar = ds[filtro & ds['level4'].str.contains('Galapagar', case=False, na=False)]

# Calcular promedios
media_valdemorillo = valdemorillo['pps'].mean()
media_galapagar = galapagar['pps'].mean()

# Resultados
print("💶 PROMEDIO PRECIO POR M²:")
print(f"• Valdemorillo: {media_valdemorillo:,.2f} €/m² (n={len(valdemorillo)} propiedades)")
print(f"• Galapagar:    {media_galapagar:,.2f} €/m² (n={len(galapagar)} propiedades)")
print(f"• Diferencia:   {abs(media_valdemorillo - media_galapagar):,.2f} €/m²")
```

#### Ejercicio 15. Traza un gráfico de barras de la mediana de los precios y explica lo que observas (debes usar el subconjunto obtenido del Ejercicio 14) (★★★)
