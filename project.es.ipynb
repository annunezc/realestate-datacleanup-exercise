#### Ejercicio 06. Elimina los NAs del dataset, si aplica (‚òÖ‚òÖ‚òÜ)

Imprima una comparaci√≥n entre las dimensiones del DataFrame original versus el DataFrame despu√©s de las eliminaciones.


```python
# Guardar dimensiones originales
dim_original = ds.shape

# Eliminar filas con NA (todos los valores faltantes)
ds_sin_na = ds.dropna()

# Guardar nuevas dimensiones
dim_sin_na = ds_sin_na.shape

# Crear DataFrame comparativo
comparacion = pd.DataFrame({
    'Dataset': ['Original', 'Sin NAs'],
    'Filas': [dim_original[0], dim_sin_na[0]],
    'Columnas': [dim_original[1], dim_sin_na[1]],
    'Filas eliminadas': [0, dim_original[0] - dim_sin_na[0]]
})

# Mostrar resultados
print("Comparaci√≥n de dimensiones:")
print(comparacion.to_string(index=False))

# Mostrar porcentaje de datos perdidos
porcentaje_perdido = ((dim_original[0] - dim_sin_na[0]) / dim_original[0]) * 100
print(f"\nSe elimin√≥ el {porcentaje_perdido:.2f}% de las filas originales")
```

#### Ejercicio 07. ¬øCu√°l la media de precios en la poblaci√≥n (columna level5) de "Arroyomolinos (Madrid)"? (‚òÖ‚òÖ‚òÜ)

Imprima el valor obtenido.

```python
## Filtrar propiedades en Arroyomolinos (sin NA y precio > 0)
filtro = (
    ds['level4'].str.contains('Arroyomolinos', case=False, na=False) & 
    (ds['price'] > 0)
)
arroyomolinos_data = ds[filtro]

# Calcular estad√≠sticas
if not arroyomolinos_data.empty:
    media_precio = arroyomolinos_data['price'].mean()
    cantidad_propiedades = len(arroyomolinos_data)
    
    print(f"üîç An√°lisis para: Arroyomolinos (Madrid) (excluyendo precio 0)")
    print(f"‚Ä¢ Propiedades v√°lidas encontradas: {cantidad_propiedades}")
    print(f"‚Ä¢ Precio promedio: {media_precio:,.2f} ‚Ç¨")
    print(f"‚Ä¢ Rango de precios v√°lidos: {arroyomolinos_data['price'].min():,.2f} ‚Ç¨ - {arroyomolinos_data['price'].max():,.2f} ‚Ç¨")
    
    # Propiedades m√°s econ√≥micas (excluyendo 0)
    print("\nüí∏ 3 propiedades m√°s econ√≥micas (con precio > 0):")
    print(arroyomolinos_data.nsmallest(3, 'price')[['address', 'price', 'surface', 'rooms']].to_string(index=False))
    
else:
    print("No se encontraron propiedades v√°lidas (precio > 0) en Arroyomolinos (Madrid)")
```

#### Ejercicio 08. Trazar el histograma de los precios para la poblaci√≥n (level5 column) de "Arroyomolinos (Madrid)" y explica qu√© observas (‚òÖ‚òÖ‚òÜ)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve an√°lisis del trazado.


```python
import matplotlib.pyplot as plt
# Filtrar propiedades en Arroyomolinos con precio > 0
filtro = (
    ds['level4'].str.contains('Arroyomolinos', case=False, na=False) & 
    (ds['price'] > 0)
)
arroyomolinos_data = ds[filtro]

# Crear histograma
plt.figure(figsize=(12, 6))
plt.hist(arroyomolinos_data['price'], 
         bins=20, 
         color='#4CAF50', 
         edgecolor='darkgreen',
         alpha=0.7)

# Personalizaci√≥n
plt.title('Distribuci√≥n de Precios en Arroyomolinos (Madrid)', fontsize=15, pad=20)
plt.xlabel('Precio (‚Ç¨)', fontsize=12)
plt.ylabel('N√∫mero de Propiedades', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# A√±adir l√≠nea de media
media = arroyomolinos_data['price'].mean()
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, 
            label=f'Media: {media:,.2f} ‚Ç¨')

# Formatear ejes
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f} ‚Ç¨'))
plt.xticks(rotation=45)

plt.legend()
plt.tight_layout()
plt.show()

# Estad√≠sticas complementarias
print(f"\nüìä Estad√≠sticas clave de precios en Arroyomolinos (n={len(arroyomolinos_data)} propiedades):")
print(f"‚Ä¢ Media: {media:,.2f} ‚Ç¨")
print(f"‚Ä¢ Mediana: {arroyomolinos_data['price'].median():,.2f} ‚Ç¨")
print(f"‚Ä¢ Rango: {arroyomolinos_data['price'].min():,.2f} ‚Ç¨ - {arroyomolinos_data['price'].max():,.2f} ‚Ç¨")
```
